```python
df["genres"] = df["genres"].apply(lambda x:[i.replace(" ","") for i in x])
df["keywords"] = df["keywords"].apply(lambda x:[i.replace(" ","") for i in x])
# df["tagline"] = df["tagline"].apply(lambda x:[i.replace(" ","") for i in x])
```

```python
df["tags"] =  df["genres"] + df["keywords"] +df["tagline"] + df["overview"]
```

```python
df = df[["id","tags","original_title"]]
df["original_title"] = df["original_title"].str.lower()
df["tags"] = df["tags"].apply(lambda x: [i.lower() for i in x])


df
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
def stemm(text):
    y = []
    
    if isinstance(text, list):
        words = text
    else:
        words = text.split()
    
    for i in words:
        y.append(ps.stem(i))
        
    return " ".join(y)

df["tags"] = df["tags"].apply(stemm)
```

```python
df["tags"][0]
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer


cv = TfidfVectorizer(
    max_features=50000,
    stop_words="english"
)

vectors = cv.fit_transform(df["tags"]).toarray()
vectors
```

```python
from sklearn.metrics.pairwise import cosine_similarity

similarity = cosine_similarity(vectors)
similarity
```

```python

def recommend(movie_name):
    movie_name = movie_name.lower()

    matches = df[df["original_title"] == movie_name]

    if matches.empty:
        return "Movie not found"

    movie_index = matches.index[0]
    distance = similarity[movie_index]

    recommended_movies = sorted(
        list(enumerate(distance)),
        reverse=True,
        key=lambda x: x[1]
    )[1:6]

    return [df.iloc[i[0]]["original_title"] for i in recommended_movies]
```

```python
sorted(list(enumerate(similarity[0])),reverse=True,key=lambda x:x[1])[1:6]
```
